```python
df_flights['month'] = pd.DatetimeIndex(df_flights['fl_date']).month
```

```python
filter = df_flights[['month','flights']]
flights_per_month = filter.groupby('month', as_index=False).sum('flights').sort_values(by='month', ascending=False)
```

```python
g = sns.lineplot(data=flights_per_month, x='month',y='flights')
```
