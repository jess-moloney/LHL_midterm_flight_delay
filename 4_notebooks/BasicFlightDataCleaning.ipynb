```python
df_flight = pd.read_csv('2_data/flights_random_sample.csv')
df_pass = pd.read_csv('2_data/passenger_random_sample.csv')
df_fuel = pd.read_csv('2_data/fuel_consumption.csv')

dfs = [df_flight, df_pass, df_fuel]
```

```python
for df in dfs:
    df.dropna(axis = 1, how = 'all', inplace = True)
```

```python
df_f = df_flight[['fl_date','mkt_carrier',
'mkt_carrier_fl_num','op_unique_carrier','tail_num','op_carrier_fl_num',
'origin_airport_id','dest_airport_id','crs_dep_time','crs_arr_time',
'crs_elapsed_time','distance','arr_delay']].copy()
```

```python
df_f.dropna(axis = 0, how = 'any', inplace=True)
df_f.info()
```

```python
df_f
```

```python
df_f.T.drop_duplicates().T
df_f.drop('op_unique_carrier', axis = 1, inplace = True)
```

```python
df_f[['year', 'month', 'day']] = df_f['fl_date'].str.split('-', expand=True).astype('int')
df_f.tail_num = df_f.tail_num.astype('category')
df_f.mkt_carrier = df_f.mkt_carrier.astype('category')
df_f.tail_num = df_f.tail_num.cat.codes
df_f.mkt_carrier = df_f.mkt_carrier.cat.codes
df_f.drop('fl_date', axis = 1, inplace = True)
df_f.dtypes
```

```python
df_f.to_csv('2_data/baseline_data.csv')
```
