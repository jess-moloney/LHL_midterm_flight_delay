### Transform Categorical Variables into Numeric Data

```python
nom_feats_train = df_flights_train.dtypes[(df_flights_train.dtypes == 'object') | (df_flights_train.dtypes == 'category') ].index.tolist()
bool_feats_train = df_flights_train.dtypes[df_flights_train.dtypes == 'bool'].index.tolist()

cat_feats_train = nom_feats_train + bool_feats_train

df_numeric_train = df_flights_train.drop(columns=cat_feats_train)
df_numeric_train = df_numeric_train.drop(columns='arr_delay')

df_dummy_train = pd.get_dummies(df_flights_train[nom_feats_train])
df_dummy_train.head()
```

```python
nom_feats_test = df_flights_test.dtypes[(df_flights_test.dtypes == 'object') | (df_flights_test.dtypes == 'category') ].index.tolist()
bool_feats_test = df_flights_test.dtypes[df_flights_test.dtypes == 'bool'].index.tolist()

cat_feats_test = nom_feats_test + bool_feats_test

df_numeric_test = df_flights_test.drop(columns=cat_feats_test)
df_numeric_test = df_numeric_test.drop(columns='arr_delay')

df_dummy_test = pd.get_dummies(df_flights_test[nom_feats_test])
df_dummy_test.head() #drop_first=True?
```
